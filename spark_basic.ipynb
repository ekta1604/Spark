1. ##### select() - select columns from dataframe
2. ##### collect() - Retrive data from dataframe 
    - select() is a transformation that returns a new DataFrame and holds the columns that are selected whereas collect() is an action that returns the entire data set in an Array to the driver.
3. ##### filter() and where() - filter() function is used to filter the rows from RDD/DataFrame based on the given condition or SQL expression, you can also use where() clause instead of the filter() if you are coming from an SQL background, both these functions operate exactly the same.
4. ##### distinct() - drop/remove the duplicate rows (all columns) from DataFrame
5. ##### dropDuplicates() - drop rows based on selected (one or multiple) columns.
6. ##### sort() or orderBy() - sort DataFrame by ascending or descending order based on single or multiple columns
7. ##### groupBy() - collect the identical data into groups on DataFrame and perform count, sum, avg, min, max functions on the grouped data
8. ##### join() - combine two DataFrames and by chaining these you can join multiple DataFrames; it supports all basic join type operations available in traditional SQL like INNER, LEFT OUTER, RIGHT OUTER, LEFT ANTI, LEFT SEMI, CROSS, SELF JOIN
    - join(self, other, on=None, how=None)
    - join() operation takes parameters as below and returns DataFrame.
    - param other: Right side of the join
    - param on: a string for the join column name
    - param how: default inner. Must be one of inner, cross, outer,full, full_outer, left, left_outer, right, right_outer,left_semi, and left_anti.
    - https://sparkbyexamples.com/pyspark/pyspark-join-explained-with-examples/ 
9. ##### Aggrigate Functions 
      - **approx_count_distinct()** function returns the count of distinct items in a group.
      - **avg()** function returns the average of values in the input column.
      - **collect_list()** function returns all values from an input column with duplicates.
      - **countDistinct()** function returns the number of distinct elements in a columns.
      - **count()** function returns number of elements in a column.
      - **sumDistinct()** function returns the sum of all distinct values in a column.
10. ##### Window Functions - To perform an operation on a group first, we need to partition the data using Window.partitionBy() , and for row number and rank function we need to additionally order by on partition data using orderBy clause.
      - **row_number()** window function is used to give the sequential row number starting from 1 to the result of each window partition.
      - **rank()** window function is used to provide a rank to the result within a window partition. This function leaves gaps in rank when there are ties.
      - **dense_rank()** window function is used to get the result with rank of rows within a window partition without any gaps. This is similar to rank() function difference being rank function leaves gaps in rank when there are ties.


11. ##### PySpark Built-In Function 
      - when() is a SQL function that returns a Column type and otherwise() is a function of Column, if otherwise() is not used, it returns a None/NULL value.
      - when() is SQL function, in order to use this first you should import and this returns a Column type, otherwise() is a function of Column, when otherwise() not used and none of the conditions met it assigns None (Null) value. Usage would be like when(condition).otherwise(default).
      - when() function take 2 parameters, first param takes a condition and second takes a literal value or Column, if condition evaluates to true then it returns a value from second param.

```
# Select
df = spark.read.csv("/FileStore/tables/cricket/335982.csv",header=True)
# Select single or multiple comlumns 
df.select("batter","runs_batter").show()
df.select(df.batter,df.runs_batter).show()
df.select(df["batter"],df["runs_batter"]).show()

# By using col function 
from pyspark.sql.functions import col
df.select(col("batter"),col("runs_batter")).show()

#Select all columns form list 
df.select([col for col in df.columns]).show()
df.select("*").show()

#Select columns by index
#Selects first 3 columns and top 3 rows
df.select(df.columns[:3]).show(3)

#Selects columns 2 to 4  and top 3 rows
df.select(df.columns[2:4]).show(3)
```

```
# Collect
df = spark.read.csv("/FileStore/tables/cricket/335982.csv",header=True)
data = df.collect()
print(data) # collect() do not return dataframe, returns an array

for row in data:
    print(row['season'] + "," +str(row['city'])) #only season and city
    
data1 = df.collect()[0] #First row 
print(data1)
```

```
#Filter()
df = spark.read.csv("/FileStore/tables/cricket/335982.csv",header=True)

#Using equal sign 
df.filter(df.kind=="caught").show(truncate=False)

#Using SQL col() function
from pyspark.sql.functions import col
df.filter(col("kind") == "caught") \
    .show(truncate=False)

#Using SQL Expression
df.filter("kind == 'caught'").show()
#For not equal
df.filter("kind != 'caught'").show()
df.filter("kind <> 'caught'").show()

#Filter multiple condition
df.filter( (df.kind  == "caught") & (df.gender  == "male") ) \
    .show(truncate=False)
```

```
# Drop Duplicate 
#Get Distinct Rows (By Comparing All Columns)
df = spark.read.csv("/FileStore/tables/cricket/335982.csv",header=True)
distinctDF = df.distinct()
print("Distinct count: "+str(distinctDF.count()))
distinctDF.show(truncate=False)

df2 = df.dropDuplicates()
print("Distinct count: "+str(df2.count()))
df2.show(truncate=False)

#Multiple
dropDisDF = df.dropDuplicates(["batter","kind"])
print("Distinct count of batter & kind : "+str(dropDisDF.count()))
dropDisDF.show(truncate=False)
```

```
# Sort 
df = spark.read.csv("/FileStore/tables/cricket/335982.csv",header=True)

df.orderBy("runs_batter","batter").show(truncate=False)
df.orderBy(col("runs_batter"),col("batter")).show(truncate=False)

#Sort by Ascending (ASC) and descending(DESC)
df.sort(df.runs_batter.asc(),df.batter.asc()).show(truncate=False)
df.sort(col("runs_batter").asc(),col("batter").asc()).show(truncate=False)
df.orderBy(col("runs_batter").asc(),col("batter").asc()).show(truncate=False)
```

```
#GroupBy
df = spark.read.csv("/FileStore/tables/cricket/335982.csv",header=True)
from pyspark.sql.functions import sum,avg,max,count,min,mean
df.groupBy("team").count().show()
df.groupBy("team").sum("runs_batter").show(truncate=False)
df.groupBy("team").min("runs_batter").show(truncate=False)
df.groupBy("team").max("runs_batter").show(truncate=False)
df.groupBy("team").avg("runs_batter").show(truncate=False)
df.groupBy("team").mean("runs_batter").show(truncate=False)

#GroupBy on multiple columns
df.groupBy("team","batter") \
    .sum("runs_batter","runs_total") \
    .show(false)

#Running more aggregates at a time

batter_stats = df.groupBy("team", "batter").agg(
    sum("runs_batter").alias("total_runs"),
    count("ballno").alias("total_balls")
)
```

```
from pyspark.sql.functions import approx_count_distinct,avg,collect_list
df = spark.read.csv("/FileStore/tables/cricket/335982.csv",header=True)
#approx_count_distinct()
print("approx_count_distinct: " + \
      str(df.select(approx_count_distinct("city")).collect()[0][0]))

#avg (average) Aggregate Function
print("avg: " + str(df.select(avg("runs_total")).collect()[0][0]))

#collect_list Aggregate Function
df.select(collect_list("runs_batter")).show(truncate=False)





```

```
#Row_number function
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number
df=spark.read.csv("/FileStore/tables/cricket/335982.csv",header=True)
windowSpec  = Window.partitionBy("team").orderBy("runs_batter")
df.withColumn("row_number",row_number().over(windowSpec)).shw(truncate=False)
    
```

```
#rank() function
from pyspark.sql.window import Window
from pyspark.sql.functions import rank
df=spark.read.csv("/FileStore/tables/cricket/335982.csv",header=True)
windowSpec  = Window.partitionBy("team").orderBy("runs_batter")
df.withColumn("rank",rank().over(windowSpec)).show()
```

```
#when() 
from pyspark.sql.functions import when
df=spark.read.csv("/FileStore/tables/cricket/335982.csv",header=True)
df2 = df.withColumn("Wickets", when(df.kind != "-","Out").otherwise(df.kind))
df2.show()
```
